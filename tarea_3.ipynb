Importar librerias

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Cargando datos

```python
partidos=pd.read_csv("https://raw.githubusercontent.com/cienciadedatos/datos-de-miercoles/master/datos/2019/2019-04-10/partidos.txt",delimiter='\t')
```

Seleccionar datos de España y agrupar por año para obtener el total de goles

```python
españa= partidos[(partidos["equipo_1"] == 'España')][["anfitrion", "equipo_1_final"]]
españa=españa.groupby("anfitrion")[["equipo_1_final"]].sum().reset_index()
españa
```

Seleccionar datos de Argentina y agrupar por año para obtener el total de goles

```python
argentina= partidos[(partidos["equipo_1"] == 'Argentina')][["anfitrion", "equipo_1_final"]]
argentina=argentina.groupby("anfitrion")[["equipo_1_final"]].sum().reset_index()
argentina
```

Gráfico de España

```python
plt.bar(españa["anfitrion"],españa["equipo_1_final"], color="aquamarine")
plt.xlabel("Anfitrion")
plt.ylabel("Goles anotados")
plt.title("Goles anotados de España en los mundiales")
plt.xticks(rotation=90)
```

Gráfico de Argentina

```python
plt.bar(argentina["anfitrion"],argentina["equipo_1_final"], color="burlywood")
plt.xlabel("Anfitrion")
plt.ylabel("Goles anotados")
plt.title("Goles anotados de Argentina en los mundiales")
plt.xticks(rotation=90)
```

Sub-gráfico

```python
fig = plt.figure()
eje = fig.add_subplot(1, 2, 1)
plt.bar(españa["anfitrion"],españa["equipo_1_final"], color="aquamarine")
eje2 = fig.add_subplot(1, 2, 2, sharey = eje)
plt.bar(argentina["anfitrion"],argentina["equipo_1_final"], color="burlywood")
eje.set_xlabel('Anfitrión')
eje2.set_xlabel('Anfitrión')
eje.set_ylabel('Goles anotados')
plt.suptitle("Goles anotados por España y Argentina en los mundiales")
eje.set_xticklabels(españa["anfitrion"], rotation=90)
eje2.set_xticklabels(argentina["anfitrion"], rotation=90)
```

Guardando figura

```python
plt.savefig("fig", dpi=300, bbox_inches='tight')
```
